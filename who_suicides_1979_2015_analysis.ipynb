# WHO Suicide Statistics Analysis for years 1979 - 2015

Suicide is a big and serious problem. In 2016 WHO reported that we are losing someone approximately every 40 seconds and predicts that it will get much worse.

Dataset used here is created by Kamil Szymanski and is accessible on [Kaggle](https://www.kaggle.com/szamil/who-suicide-statistics). Data itself belongs to [World Health Organization](https://www.who.int/).

This analysis was done during [Data Analysis with Python: Zero to Pandas](https://jovian.ml/learn/data-analysis-with-python-zero-to-pandas) Course.

# Data Preparation

```python
# Installing libraries
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
data = pd.read_csv("who_suicide_statistics.csv")
# Removing data for 2016 because there were very few countries with data for that year
data = data.drop(data[data.year == 2016].index)
# Renaming some columns
data.rename({"sex": "gender", "suicides_no": "suicides"}, inplace = True, axis = 1)

data.head(10)
```

```python
data.columns
```

# Data Analysis

```python
# Sum of suicides throughout the years 1979-2015 by country
data_sum = data.groupby("country")[["suicides"]].sum()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data_sum)
```

```python
# Top 10 countries with the most suicides
top10_sum_country = (data_sum).sort_values(by="suicides", ascending=False).head(10)
top10_sum_country
```

```python
# Top 10 countries with least suicides
least10_sum_country = (data_sum).sort_values(by="suicides", ascending=True).head(10)
least10_sum_country
```

```python
# Trend in suicide amount
trend_sui_years = data.groupby("year")[["suicides"]].sum().sort_values(by="year", ascending=True)
trend_sui_years.head(10)
```

```python
# Top 10 of deadliest years
top10_sum_years = data.groupby("year")[["suicides"]].sum().sort_values(by="suicides", ascending=False).head(10)
top10_sum_years
```

```python
# Top 10 of least deadly years
least10_sum_years = data.groupby("year")[["suicides"]].sum().sort_values(by="suicides", ascending=True).head(10)
least10_sum_years
```

```python
# Gender distribution of suicides
gender_distribution_sum = data.groupby("gender")[["suicides"]].sum().sort_values(by="suicides")
gender_distribution_sum
```

```python
# ru by year
ru_sum_year = data.loc[data["country"] == "Russian Federation"].groupby("year")[["suicides"]].sum()
ru_sum_year
```

```python
# ru by age group
ru_sum_agegroup = data.loc[data["country"] == "Russian Federation"].groupby("age")[["suicides"]].sum()
ru_sum_agegroup
```

```python
#usa by year
usa_sum_year = data.loc[data["country"] == "United States of America"].groupby("year")[["suicides"]].sum()
usa_sum_year
```

```python
#usa by age group
usa_sum_agegroup = data.loc[data["country"] == "United States of America"].groupby("age")[["suicides"]].sum()
usa_sum_agegroup
```

```python
#japan by year
japan_sum_year = data.loc[data["country"] == "Japan"].groupby("year")[["suicides"]].sum()
japan_sum_year
```

```python
# japan by age group
japan_sum_agegroup = data.loc[data["country"] == "Japan"].groupby("age")[["suicides"]].sum()
japan_sum_agegroup
```

# Data Visualisation

```python
plt.figure(figsize=(12,8))
plt.title("Top 10 countries with most suicides")
plt.xticks(rotation=90)
sns.barplot(x = top10_sum_country.index, y = top10_sum_country.suicides, palette="rocket")
plt.ylabel("Suicides in millions")
plt.xlabel("Country")
```

```python
plt.figure(figsize=(12,8))
plt.title("Top 10 countries with least suicides")
plt.xticks(rotation=90)
sns.barplot(x = least10_sum_country.index, y = least10_sum_country.suicides, palette="rocket_r")
plt.ylabel("Suicides")
plt.xlabel("Country")
```

```python
plt.figure(figsize=(12,6))
plt.plot(trend_sui_years, color = "red")
plt.title("Suicides throughout the years 1979-2015")
plt.xlabel("Year")
plt.ylabel("Suicides")
```

```python
plt.figure(figsize=(12,8))
plt.title("Top 10 years with most suicides")
sns.barplot(x = top10_sum_years.index, y = top10_sum_years.suicides, order=top10_sum_years.sort_values("suicides", ascending=False).index, palette="Oranges_r")
plt.ylabel("Suicides")
plt.xlabel("Year")
```

```python
plt.figure(figsize=(12,8))
plt.title("Top 10 years with least suicides")
sns.barplot(x = least10_sum_years.index, y = least10_sum_years.suicides, order=least10_sum_years.sort_values("suicides", ascending=True).index, palette="Oranges")
plt.ylabel("Suicides")
plt.xlabel("Year")
```

```python
plt.figure(figsize=(12,6))
plt.title("Suicides distribution by gender")
plt.pie(gender_distribution_sum.suicides, labels = ["Female", "Male"], colors = ["pink", "grey"], explode =(0,0.1), autopct='%1.1f%%')
```

# Questions

## What are the countries with the most suicides?

Russian Federation, USA and Japan take top 3 spots and by a lot. Fourth country which is France has "only" half of Japan's suicides.

```python
plt.figure(figsize=(12,8))
plt.title("Top 10 countries with most suicides")
plt.xticks(rotation=90)
sns.barplot(x = top10_sum_country.index, y = top10_sum_country.suicides, palette="rocket")
plt.ylabel("Suicides in millions")
plt.xlabel("Country")
```

## What was the deadliest year?

2003 was the deadliest year, closely followed by 2002, 1999 and 2000.

```python
top10_sum_years.head(5)
```

## What age group commits suicide the most?

It is by far the 35-54 years group with almost 3 million suicides, followed by 55-74 years with about 2 million suicides and 25-34 years with about 1,4 million. 

```python
sum_agegroup = data.groupby("age")[["suicides"]].sum()
plt.figure(figsize=(12,8))
plt.title("Suicides distribution by age group")
sns.barplot(x = sum_agegroup.index, y = sum_agegroup.suicides, palette="plasma")
plt.ylabel("Suicides")
plt.xlabel("Age group")
```

## How does the ratio of suicides between genders looks like?

Males commit suicide about 4 times more often than females. With distribution of 76,3% to 23,7%.

```python
plt.figure(figsize=(12,6))
plt.title("Suicides distribution by gender")
plt.pie(gender_distribution_sum.suicides, labels = ["Female", "Male"], colors = ["pink", "grey"], explode =(0,0.1), autopct='%1.1f%%')
```

## How does the trend of suicides looks like?

Amount of suicides was constantly growing till 2003 and since then it started to go down. Which is interesting because WHO predicts that it should be growing again.

```python
plt.figure(figsize=(12,6))
plt.plot(trend_sui_years, color = "red")
plt.title("Suicides throughout the years 1979-2015")
plt.xlabel("Year")
plt.ylabel("Suicides")
```

# Conclusion and facts
- There was a big dip in amount of suicides in 1987-1989.
- Males commit suicide 4 times more often than females.
- Russian Federation has the biggest amount of suicides.
- Suicide is the most common in age group of 35-54 and least common in age group 5-14.
